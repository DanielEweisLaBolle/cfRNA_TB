# Figure #1B

```r
#Read in Files 
decon = read.delim("../Files_For_Manuscript/discovery_cell_types.tsv")
metadata_train = read.delim("../Files_For_Manuscript/metadata.train.tsv")
metadata_test =read.delim("../Files_For_Manuscript/metadata.test.tsv")
metadata_val =  read.delim("../Files_For_Manuscript/metadata.val.tsv")
meta_decon = rbind(metadata_train, metadata_test, metadata_val)
```

```r
## Plot Cell Types ##
# Set var for deconvolution metadata
meta_data = meta_decon
# Rename variables for legibility
decon_melt = melt(data.table(decon), id.vars = "sample_id", variable.name = "celltype", value.name = "fraction")
# Merge metadata with cell type fraction data
decon_melt = merge(decon_melt, meta_data, by.x="sample_id", by.y="sample_id")
# Rename var
results = decon_melt
# Calculate mean for each cell type across all samples
sum = results %>% group_by(celltype) %>% summarize(mean=mean(fraction))
sum = sum[order(sum$mean, decreasing=TRUE),]

genes = sum$celltype

num_cells = length(genes)
# Set color palette
pal = c('#1362AB','#740C33','#2C8DAF','#FB927F','#5CFBF4','#0234C1','#86C35E','#B414A8','#F191B8','#949D0A','#700862','#F5DB9E','#209510','#DBB008','#5C6BFA','#B27DF2','#D6F306','#5EFE36','#A9DAC9','#B2040D','#203760','#4D5D0D','#4A1E80','#F63DFA','#F8169B','#175738','#0C8559','#A1506B','#861FF1','#2BBBB3','#5E210C','#36B976','#CAA376','#D75E16','#73480D','#39F9AD','#85D2F8','#65A2FC','#FC092B','#94A514','#AEBBE6')

colors = pal
labels = genes

val_vector = colors
names(val_vector) = labels
# Manually assign colors to relevant cell types of interest
val_vector[names(val_vector) %in% c("Platelet","Myeloid.progenitor","B.cell","Endothelial.cell","NK.Cell","Monocyte","neutrophil")]= c("#72D2D3","#FFBC42","#D81159","#218380","#8F2D56","#2E294E","#CADB7A")
# All other cell types go to "Other" category and get a single color
val_vector[!(names(val_vector) %in% c("Platelet","Myeloid.progenitor","B.cell","Endothelial.cell","NK.Cell","Monocyte","neutrophil"))] =" #EBEBEB"

results$celltype2 = as.character(results$celltype)
results$celltype2[!(results$celltype2) %in% c("Platelet","Myeloid.progenitor","B.cell","Endothelial.cell","NK.Cell","Monocyte","neutrophil")] = "other"


p = results %>% 
    group_by(sample_id, cohort, celltype2) %>% 
    summarise(Fraction = sum(fraction)) %>% 
    group_by(celltype2, cohort) %>% 
    summarise(Fraction = mean(Fraction)) %>% 
    mutate(celltype2 = factor(celltype2, levels = c("Platelet", "Myeloid.progenitor", "B.cell", "Endothelial.cell", "NK.Cell", "Monocyte", "neutrophil", "other"))) %>%
    ggplot(aes(x = cohort, y = Fraction, fill = celltype2)) +
    geom_bar(position = "stack", stat = "identity", colour = "black", size = 0.15) +
    theme_minimal(base_size = 20) +
    theme(
        plot.margin = margin(1, 20, 1, 1, "mm"), 
        axis.title.x = element_blank(), 
        axis.title.y = element_text(size = 24.5, color = "black"),
        axis.text.x = element_text(size = 22.5, color = "black"),  # X-axis text
        axis.text.y = element_text(size = 22.5, color = "black"),  # Increased Y-axis label text size
        legend.position = "top",
        legend.text = element_text(size = 15.5), 
        legend.title = element_blank(), 
        legend.key.size = unit(0.25, 'cm'), 
        panel.border = element_rect(colour = "black", fill = NA, size = 2.85)
    ) +
    scale_fill_manual(values = c(val_vector[1:15], "other" = "#fff2d5"), 
                      labels = c("Platelet", "Myeloid progenitor", "B Cell", "Endothelial cell", "NK Cell", "Monocyte", "neutrophil", "other")) + 
    scale_x_discrete(labels = c("Cohort 1\nn=93", "Cohort 2\nn=98", "Cohort 3\nn=60"))

p

ggsave("../Figure_Output/Figure_1/Figure_1B.png", plot = p, width = 7.2, height = 7, dpi = 300)
```

# Figure 1C

```r
#Set Seed 
set.seed(87)
# Read in Metadata
metadata_train = read.delim("../Files_For_Manuscript/metadata.train.tsv")
metadata_test = read.delim("../Files_For_Manuscript/metadata.test.tsv")
metadata_val = read.delim("../Files_For_Manuscript/metadata.val.tsv")

#Concatenate Metadata
metadata_file_modif = rbind(metadata_train, metadata_test, metadata_val)
#Recode site to match the country it was collected from across cohorts
metadata_file_modif <- metadata_file_modif %>%
  mutate(site = recode(site, 
                       "Kisenyi" = "Uganda", 
                       "Mulago" = "Uganda"))
metadata_file_modif <- metadata_file_modif %>%
  mutate(site = ifelse(is.na(site) | site == "", "Uganda", site))

#Assign variables for differential abundance analysis
SAMPLE_ID_VAR = "sample_id"
COMP_VAR = "tb"
GROUPS = c("positive","negative")

gene_name_key = "../Files_For_Manuscript/gencode.biotype.name.key.tsv"

DESIGN = paste0("~",COMP_VAR)

#Read in the count matrices
counts_train = read.delim("../Files_For_Manuscript/counts.train.tsv",row.names=1)
counts_test = read.delim("../Files_For_Manuscript/counts.test.tsv", row.names = 1)
counts_val = read.delim("../Files_For_Manuscript/counts.val.tsv",row.names=1)
#Concatenate count matrices
counts <- cbind(counts_train, counts_test, counts_val)
samples <- metadata_file_modif %>% 
     filter(.data[[COMP_VAR]] %in% GROUPS)
rownames(samples) <- samples$sample_id
counts <- counts[, rownames(samples)]
```

## DESEQ2

```r
##------------------------------------
# Contstruct DESeq Data Set
dds <- DESeqDataSetFromMatrix(counts,
                                colData = samples,
                                design = formula(DESIGN))


##------------------------------------
# Add Gene metadata
annotation = fread(file=gene_name_key)
annotation <- annotation[match(rownames(dds), annotation$gene_id),]
all(rownames(dds) == annotation$ftcount_id)
mcols(dds) <- cbind(mcols(dds), annotation)


##------------------------------------
# Re-factor
dds[[COMP_VAR]] <- factor(dds[[COMP_VAR]], levels = GROUPS)


##------------------------------------
# DAA for all 251 TB samples
dds <- DESeq(dds)
```

```r
res <- results(dds,alpha=0.05,contrast = c(COMP_VAR,GROUPS[1], GROUPS[2]))

res$gene_name <- mcols(dds)$gene_name
res$gene_type <- mcols(dds)$gene_type

summary(res)
```

## Heatmap

```r
options(repr.plot.width=15,repr.plot.height=7)
mycolors <- list(
   tb = c("positive" = "#ff70a6",
                 "negative" = "#2E294E"),
    hiv = c("Positive"= "#f77f00",
                   "Negative" = "#8F2D56"),
    cohort = c("GHL" = "#73D2DE",
               "R2D2" = "#b5179e",
               "TBSQUARE"="gold"),
    site = c("Uganda" = "#218380",
                "Vietnam" = "#FFBC42",
                "Philippines" = "#D81159"))
       
vars = c("tb", "hiv", "cohort", "site")


htmap <- make_heatmap(res, counts, samples, "sample_id", mycolors, vars, 
                      SIG_THRESH = 0.05, FC_THRESH = 0)

options(repr.plot.width=7,repr.plot.height=7)

htmap
ggsave("../Figure_Output/Figure_1/Figure_1C.png", plot = htmap, width = 7, height = 5.7, dpi = 300)
```

```r
# Read in the IPA output file
canonical_path = read.table("../Files_For_Manuscript/ipa_output.txt" , sep="\t", header=T)

# Quantify the number of molecules in each identified pathway
canonical_path$mol_num= unlist(lapply(unlist(canonical_path$Molecules), function(x){length(unlist(strsplit( x, ",")))}))

p = canonical_path %>% 
    # Set cutoffs for Z-score, p-value, and molecule number 
    filter(abs(X.log.p.value.) > 1) %>% 
    filter(abs(z.score) > 2.55) %>% 
    filter(mol_num > 6) %>% 
    # Plot
    ggplot(aes(x = reorder(Ingenuity.Canonical.Pathways, abs(z.score)), y = abs(z.score), fill = ifelse(z.score < 0, "#D81159", "#218380"))) + 
    geom_bar(stat = 'identity') + 
    geom_text(aes(label = Ingenuity.Canonical.Pathways, y = abs(z.score) + 0.06), hjust = 0, color = "black", size = 6.5) +
    geom_blank(aes(y = max(abs(z.score)) * 4)) +  # Invisible layer to extend plot area
    coord_flip() + 
    theme_bw(base_size = 8) + 
    theme(
        axis.title.y = element_blank(),
        axis.text.y = element_blank(),
        axis.text.x = element_text(size = 20),
        axis.line.x = element_line(color = "black", size = 0.5),
        legend.key.size = unit(0.25, 'cm'), 
        legend.title = element_text(size = 5), 
        legend.text = element_text(size = 4), 
        legend.position = "none", 
        panel.border = element_blank(),
        panel.grid.major = element_blank(),
        panel.grid.minor = element_blank(),
        plot.margin = margin(5.5, 5.5, 5.5, 5.5, "points")
    ) + 
    ylab("z-score")

p
ggsave("../Figure_Output/Figure_1/Figure1_D.png", plot = p, width = 6, height = 6, dpi = 300)
ggsave("../Figure_Output/Figure_1/Figure1_D.pdf", plot = p, width = 6, height = 6, device = "pdf")
```

```r
# Assuming auc_all is your dataframe and long_data is prepared as shown
library(tidyverse)

#Read in the model performance results
auc_all <- read.table("../Files_For_Manuscript/ml_auc.txt", header = TRUE)

# Filter for only train and test methods
auc_filtered <- auc_all %>%
  filter(method %in% c("train", "test"))

# Order models by descending test AUC
models_ordered <- auc_filtered %>%
  filter(method == "test") %>%
  arrange(auc) %>%
  pull(model)

long_data <- auc_filtered %>%
  gather(key = "Metric", value = "Value", auc)

# Set the levels of the model factor in long_data according to this order
long_data$model <- factor(long_data$model, levels = models_ordered)

# Define shapes for train and test
shape_values <- c("train" = 21, "test" = 24)

# Plotting code
p <- ggplot(long_data, aes(y = model, x = Value, shape = method, color = method)) +
  geom_point(size = 4, stroke = 1.5) +
  scale_shape_manual(values = shape_values) +
  scale_color_manual(values = c("train" = "blue", "test" = "red")) +
  theme_minimal(base_size = 20) +
  theme(text = element_text(family = "sans"),
        axis.title = element_blank(), 
        axis.text.y = element_text(angle = 0, hjust = 1, vjust = 0.5, color = "black"), 
        legend.position = "top", 
        legend.text = element_text(size = 22, color = "black"),  # Increase legend text size
        legend.title = element_blank(),  # Remove legend title
        legend.key.size = unit(1.5, "lines"),  # Increase legend key size
        legend.box.margin = margin(0, 0, -22, 0),  # Adjust legend position
        axis.text.x = element_text(color = "black", size = 21),
        panel.border = element_rect(color = "black", fill = NA, size = 2)) +
  labs(x = "AUC Value", y = "Model", color = "", shape = "") +
  guides(shape = guide_legend(override.aes = list(size = 7)))  # Increase shape size in legend

# Display the plot
print(p)

# Save the plot
ggsave("../Figure_Output/Figure_1/Figure1_F.png", plot = p, width = 5.8, height = 7.42, dpi = 300)
ggsave("../Figure_Output/Figure_1/Figure1_F.pdf", plot = p, width = 5.8, height = 7.42, device = "pdf")
```
